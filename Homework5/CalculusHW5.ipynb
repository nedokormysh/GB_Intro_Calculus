__1.__ –ù–∞–π—Ç–∏ –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—É—é $y'_{x}$ —Ñ—É–Ω–∫—Ü–∏–∏:

### $$\mbox{arctg}(\frac{y}{x})=\ln \sqrt{x^2+y^2}$$

$$\Big(\mbox{arctg}(\frac{y}{x})\Big)'=\Big(\ln \sqrt{x^2+y^2}\Big)'$$

$$\frac{1}{1+(\frac{y}{x})^2}\cdot\Big(\frac{y}{x}\Big)'=\frac{1}{\sqrt{x^2+y^2}}\Big(\sqrt{x^2+y^2}\Big)'$$

$$\frac{x^2}{x^2+y^2}\cdot\Big(\frac{y'x-x'y}{x^2}\Big)=\frac{1}{\sqrt{x^2+y^2}}\frac{1}{2}\Big(({x^2+y^2})^{-\frac{1}{2}}\Big)(x^2+y^2)'$$

$$\frac{y'x-y}{x^2+y^2}=\frac{1}{\sqrt{x^2+y^2}}\frac{1}{2}\frac{1}{\sqrt{x^2+y^2}}(2x+2yy')$$

$$\frac{y'x-y}{x^2+y^2}=\frac{x+yy'}{x^2+y^2}$$
$$y'x-y=x+yy'$$
$$y'x-yy'=x+y$$
$$y'(x-y)=x+y$$
$$y'=\frac{x+y}{x-y}$$

```python
from sympy import *
init_printing()
```

```python
x,y = symbols('x y')
f = atan(y/x)-ln(sqrt(x**2+y**2))
f
```

```python
-diff(f, x)/diff(f, y)
```

__2.__ –ù–∞–π—Ç–∏ –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—É—é $y'_{x}$ —Ñ—É–Ω–∫—Ü–∏–∏:

### \begin{cases}
   y=\frac{t^2}{t-1}, \\
   x=\frac{t}{t^2-1} 
 \end{cases}


 \begin{cases}
   y'_t=\frac{2t(t-1)-t^2}{(t-1)^2}, \\
   x'_t=\frac{t^2-1-2t^2}{(t^2-1)^2} 
 \end{cases}

 \begin{cases}
   y'_t=\frac{t^2-2t}{(t-1)^2}, \\
   x'_t=\frac{-t^2-1}{(t^2-1)^2} 
 \end{cases}

 $$y'_{x}=\frac{y'_{t}}{x'_{t}}=\frac{t^2-2t}{(t-1)^2}\cdot\frac{(t^2-1)^2}{-t^2-1}=\frac{(t^2-2t)(t^2-1)(t^2-1)}{-(t-1)^2(t^2+1)}=\frac{(t^2-2t)(t-1)(t+1)(t-1)(t+1)}{-(t-1)^2(t^2+1)}=\frac{(t^2-2t)(t+1)}{-(t^2+1)}$$

```python
t=Symbol('t')
y=t/(t-1)
y
```

```python
x=t/(t**2-1)
x
```

```python
diff(y, t)/diff(x, t)
```
